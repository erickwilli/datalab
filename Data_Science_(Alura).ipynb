```python
total_por_lingua = tmdb["original_language"].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc["en"] # LOC = BUSCA O ÍNDICE
total_outros = total_geral - total_de_ingles
print(f'Total Geral: {total_geral} \n Total de Inglês: {total_de_ingles}\n Total do Resto: {total_outros}')
```

```python
# CRIANDO UM DATAFRAME
```

```python
dados = {
    "lingua" : ["ingles", "outros"],
    "total" : [total_de_ingles, total_outros]
}
dados =pd.DataFrame(dados)
dados
```

```python
sns.barplot(data=dados, x="lingua",y="total")
plt.title("Comparativo Inglês x Outros Idiomas")
plt.show()
```

```python
dados.plot(kind="pie",y="total",labels=dados["lingua"])
```
